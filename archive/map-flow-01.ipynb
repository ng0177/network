### model

```python
# roberto
n_features = x_data.shape[1]
# 32 inner nodes, 2 input nodes
# model.add(Dense(32, input_dim=2, activation='tanh'))
inputs  = keras.Input(shape=(n_features,))
dense   = keras.layers.Dense(512, activation="relu")
x       = dense(inputs)
x       = keras.layers.Dense(256, activation="relu")(x)
x       = keras.layers.Dense(128, activation="relu")(x)
x       = keras.layers.Dense(64, activation="relu")(x)
x       = keras.layers.Dense(16, activation="relu")(x)
outputs = keras.layers.Dense(1)(x)

model = keras.Model(inputs=inputs, outputs=outputs)
model.compile(optimizer='adam', loss='mse')
model.summary()
```

### train

```python
# 30,64
model.fit(x_data, y_data, batch_size=64, epochs=300)
```

### predict

```python
# from input
y_predicted = model.predict(x_data)

# plot
plt.scatter(x_data[::1,[1]], y_data[::1], s=50)
plt.plot(x_data, y_predicted, 'r', linewidth=3)
plt.grid()
#plt.show()
plt.savefig('training.png', dpi=300)
#files.download("training.png") 
```
